# Dataset and Story

MovieLens, a movie recommendation service, provided the dataset. It contains the rating scores for these movies along with the movies.

It contains 2,000,0263 ratings across 27,278 movies. This data was created by 138,493 users between 09 January 1995 and 31 March 2015. This data set was created on October 17, 2016. Users are randomly selected. It is known that all selected users voted for at least 20 movies.

**Variables of the data set:**

**movie.csv**
- movieId – Unique movie number. (UniqueID) 
- title – Movie name

**rating.csv**
* userid – Unique user number. (UniqueID) 
* movieId – Unique movie number. (UniqueID) 
* rating – The rating given to the movie by the user
* timestamp – Evaluation date

## BPR plain Model

Bayesian Personalized Ranking (BPR) is a collaborative filtering model that tries to recommend items by contrasting good (seen) item vectors with bad (unseen) item vectors based on user vector, then it recalculates all vectors with stochastic gradient descent

```python
import scipy.sparse as sparse
import implicit

train_df['user_id'] = train_df['userId'].astype('category').cat.codes
train_df['item_id'] = train_df['movieId'].astype('category').cat.codes

item_lookup = train_df[['item_id', 'movieId']].drop_duplicates()
item_lookup["item_id"] = item_lookup["item_id"].astype(np.int64)
user_lookup = train_df[['user_id', 'userId']].drop_duplicates()
user_lookup["user_id"] = user_lookup["user_id"].astype(np.int64)

weight = list(train_df.rating)
users = list(np.sort(train_df.user_id.unique()))
items = list(np.sort(train_df.item_id.unique()))

rows = train_df.user_id
cols = train_df.item_id 
data_sparse = sparse.csr_matrix((weight, (rows, cols)), shape=(len(users), len(items)))
```

```python
train_df
```

```python
modelBPR = implicit.bpr.BayesianPersonalizedRanking(learning_rate=0.1, verify_negative_samples=True, iterations=100,
                                                    factors=50)
modelBPR.fit(data_sparse.T)
```

```python
movies_for_users = modelBPR.recommend_all(data_sparse, N=50, filter_already_liked_items=True)
```

```python
test=test_df.merge(user_lookup, on='userId').merge(item_lookup, on='movieId')[['user_id', 'item_id', 'timestamp']]
test=test.sort_values(by='timestamp')[['user_id', 'item_id']].groupby('user_id').agg(list)
test
```

```python
all_sim = []
for i in tqdm(test.index):
    user_sim = []
    for k in movies_for_users[i]:
        item_sim = []
        movie_id = item_lookup.loc[item_lookup['item_id'] == k, 'movieId'].values[0]
        user_recom_item_id = df2[df2['movieId'] == movie_id].index[0]
        for j in test.loc[i, 'item_id'][-5:]:
            movie_id = item_lookup.loc[item_lookup['item_id'] == j, 'movieId'].values[0]
            user_history_item_id = df2[df2['movieId'] == movie_id].index[0]
            item_sim.append(cosine_similarities[user_history_item_id, user_recom_item_id])
        user_sim.append(sum(item_sim) / len(item_sim))
    all_sim.append(user_sim)
```

```python
len(all_sim[0])
```

```python
def sort_list(list1, list2):
 
    zipped_pairs = zip(list2, list1)
 
    z = [x for _, x in sorted(zipped_pairs, reverse=True)]
 
    return z

def metrics_at_n2(user_row, recommended_items, ground_truth_items, n):
    # Take only the top-N recommended items
    recommended_items = sort_list(recommended_items, all_sim[test.index.to_list().index(user_row)])
    
    recommended_at_n = recommended_items[:n]
    
    # Calculate the number of relevant items in the top-N recommended items
    num_relevant_at_n = len(set(recommended_at_n) & set(ground_truth_items))
    
    # Calculate precision at N
    precision_at_n = num_relevant_at_n / n if n > 0 else 0
    
    # Calculate recall at N
    recall_at_n = num_relevant_at_n / len(ground_truth_items) if len(ground_truth_items) > 0 else 0
    
    # Calculate F1-score at N
    f1_at_n = 2 * (precision_at_n * recall_at_n) / (precision_at_n + recall_at_n) if precision_at_n + recall_at_n > 0 else 0
    
    return f1_at_n, recall_at_n, precision_at_n


ns=[50, 20, 10, 5]
f1s = []
recalls = []
precisions = []
model_names = ['BPR+tfidfRank'] * 4

for n in ns:
    f1_at_n_all = []
    recall_at_n_all = []
    precision_at_n_all = []
    for user_row in test.index:
        f1_at_n, recall_at_n, precision_at_n  = metrics_at_n2(user_row, movies_for_users[user_row], test.loc[user_row].values[0], n)
        f1_at_n_all.append(f1_at_n)
        recall_at_n_all.append(recall_at_n)
        precision_at_n_all.append(precision_at_n)
    print(f"F1-score at {n}: {np.mean(f1_at_n_all)}")
    print(f"Recall at {n}: {np.mean(recall_at_n_all)}")
    print(f"Precision at {n}: {np.mean(precision_at_n_all)}")
    f1s.append(np.mean(f1_at_n_all))
    recalls.append(np.mean(recall_at_n_all))
    precisions.append(np.mean(precision_at_n_all))
```

```python
def metrics_at_n(recommended_items, ground_truth_items, n):
    # Take only the top-N recommended items
    recommended_at_n = recommended_items[:n]
    
    # Calculate the number of relevant items in the top-N recommended items
    num_relevant_at_n = len(set(recommended_at_n) & set(ground_truth_items))
    
    # Calculate precision at N
    precision_at_n = num_relevant_at_n / n if n > 0 else 0
    
    # Calculate recall at N
    recall_at_n = num_relevant_at_n / len(ground_truth_items) if len(ground_truth_items) > 0 else 0
    
    # Calculate F1-score at N
    f1_at_n = 2 * (precision_at_n * recall_at_n) / (precision_at_n + recall_at_n) if precision_at_n + recall_at_n > 0 else 0
    
    return f1_at_n, recall_at_n, precision_at_n


for n in ns:
    f1_at_n_all = []
    recall_at_n_all = []
    precision_at_n_all = []
    for user_row in test.index:
        f1_at_n, recall_at_n, precision_at_n  = metrics_at_n(movies_for_users[user_row], test.loc[user_row].values[0], n)
        f1_at_n_all.append(f1_at_n)
        recall_at_n_all.append(recall_at_n)
        precision_at_n_all.append(precision_at_n)
    print(f"F1-score at {n}: {np.mean(f1_at_n_all)}")
    print(f"Recall at {n}: {np.mean(recall_at_n_all)}")
    print(f"Precision at {n}: {np.mean(precision_at_n_all)}")
    f1s.append(np.mean(f1_at_n_all))
    recalls.append(np.mean(recall_at_n_all))
    precisions.append(np.mean(precision_at_n_all))
    model_names.append('BPR')
```

## Genre filtering

To help with resolving BPR model hallucinations we will use filtering recommended items by relevant genres

```python
train_df_genres = train_df_genres.merge(user_lookup, on='userId')

train_df_genres['genre_id'] = train_df_genres['genre'].astype('category').cat.codes
genre_lookup = train_df_genres[['genre', 'genre_id']].drop_duplicates()
genre_lookup["genre_id"] = genre_lookup["genre_id"].astype(np.int64)

train_df_genres['weight'] = 1

weight = list(train_df_genres.weight)
genres = list(np.sort(train_df_genres.genre_id.unique()))

rows = train_df_genres.user_id
cols = train_df_genres.genre_id 
data_sparse = sparse.csr_matrix((weight, (rows, cols)), shape=(len(users), len(genres)))
```

```python
model_genre = implicit.bpr.BayesianPersonalizedRanking(learning_rate=0.1, verify_negative_samples=True, iterations=100,
                                                    factors=50)
model_genre.fit(data_sparse.T)
```

```python
genres_for_users = model_genre.recommend_all(data_sparse, N=5)
```

```python
mov_gen_lookup = movie_genres.merge(item_lookup, on='movieId').merge(genre_lookup, on='genre')
mov_gen_lookup.drop(columns=['movieId', 'genre'], inplace=True)
mov_gen_lookup = mov_gen_lookup.groupby('genre_id').agg(list)
```

```python
test_genres = train_df_genres[train_df_genres.user_id.isin(test.index)][['user_id', 'genre_id']].groupby('user_id').agg(list)
```

```python
def get_top_n_vectors_by_similarity(items, vectors, query_vector, n):
    similarities = np.dot(vectors, query_vector)  # Calculate inner product similarities
    top_n_indices = np.argsort(similarities)[::-1][:n]  # Get indices of top N similar vectors
    return items[top_n_indices].tolist()
```

```python
def flatten_list(list_of_lists):
    return [item for sublist in list_of_lists for item in sublist]
```

```python
recommended_genres_hyb = []
for user in tqdm(test.index):
    recommended_g = []
    recommended_g+=test_genres.loc[user].values[0]
    recommended_g+=genres_for_users[user].tolist()
    recommended_g = get_unique_items_preserve_order(recommended_g)[:]
    
    query_items = mov_gen_lookup.loc[recommended_g].item_id.values.tolist()
    query_items = flatten_list(query_items)
    query_items = np.unique(query_items)
    query_items = np.setdiff1d(query_items, train.loc[user].values[0])
    
    recommended = get_top_n_vectors_by_similarity(query_items, modelBPR.item_factors[query_items], modelBPR.user_factors[user], 50)
    
    recommended_genres_hyb.append(recommended)
```

```python
for n in ns:
    f1_at_n_all = []
    recall_at_n_all = []
    precision_at_n_all = []
    for i,user_row in enumerate(test.index):
        f1_at_n, recall_at_n, precision_at_n  = metrics_at_n(recommended_genres_hyb[i], test.loc[user_row].values[0], n)
        f1_at_n_all.append(f1_at_n)
        recall_at_n_all.append(recall_at_n)
        precision_at_n_all.append(precision_at_n)
    print(f"F1-score at {n}: {np.mean(f1_at_n_all)}")
    print(f"Recall at {n}: {np.mean(recall_at_n_all)}")
    print(f"Precision at {n}: {np.mean(precision_at_n_all)}")
    f1s.append(np.mean(f1_at_n_all))
    recalls.append(np.mean(recall_at_n_all))
    precisions.append(np.mean(precision_at_n_all))
    model_names.append('genreFilt+BPR')
```

# Hybrid Model

Here we will use both genre filtering and similar items to enhance movie recommendations

```python
recommended_genres_sim_hyb =  []
for i,user in tqdm(enumerate(test.index)):
    recommended = []
    for item in train.loc[user][0][:1]:
        recommended+=item_similar.get(item, [])[:1]
        
    recommended+=recommended_genres_hyb[i]
    recommended = get_unique_items_preserve_order(recommended)[:50]
    recommended_genres_sim_hyb.append(recommended)
```

```python
for n in ns:
    f1_at_n_all = []
    recall_at_n_all = []
    precision_at_n_all = []
    for i,user_row in enumerate(test.index):
        f1_at_n, recall_at_n, precision_at_n  = metrics_at_n(recommended_genres_sim_hyb[i], test.loc[user_row].values[0], n)
        f1_at_n_all.append(f1_at_n)
        recall_at_n_all.append(recall_at_n)
        precision_at_n_all.append(precision_at_n)
    print(f"F1-score at {n}: {np.mean(f1_at_n_all)}")
    print(f"Recall at {n}: {np.mean(recall_at_n_all)}")
    print(f"Precision at {n}: {np.mean(precision_at_n_all)}")
    f1s.append(np.mean(f1_at_n_all))
    recalls.append(np.mean(recall_at_n_all))
    precisions.append(np.mean(precision_at_n_all))
    model_names.append('genreFilt+sim+BPR')
```
